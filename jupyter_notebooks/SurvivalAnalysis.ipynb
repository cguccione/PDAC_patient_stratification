```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from sksurv.nonparametric import kaplan_meier_estimator
from lifelines.statistics import logrank_test
```

```python
meta = pd.read_csv('TCGA_score.csv')
#display(meta.columns)
meta['id'] = meta['id'].str.replace('.01A','')
meta['id'] = meta['id'].str.replace('.','-')
meta
```

```python
df = pd.read_csv('../pre-processing/TCGA_survial_PAAD.csv')
#display(df.columns)

df = df[['submitter_id', 'days_to_death', 'days_to_last_follow_up']].copy() 
df['days_to_death'] = df['days_to_death'].fillna(0)

df['Status'] = df.apply(lambda row: bool(row.days_to_death), axis=1)

#Add survial in days based upon 
df["Survival_in_days"] = 0
for index, row in df.iterrows():
    if row['Status'] == True:
        df.at[index,'Survival_in_days'] = row['days_to_death']
    else:
        df.at[index,'Survival_in_days'] = row['days_to_last_follow_up']

df = df.drop(['days_to_death', 'days_to_last_follow_up'], axis = 1)

df.to_csv('TCGA_survial_survival.csv')
        
df
```

### Splitting data into groups

```python
#immune active vs. other (based on cd8 score) 

cd8_low = pd.DataFrame()
cd8_high = pd.DataFrame()

for index, row in df.iterrows():
    
    
    if row['submitter_id'] in list(meta['id']):
        current_df = meta[meta['id'] == row['submitter_id']]
        ###There are some things in df(we pulled from TCGA) which don't have matches in our acutal dataset

        if float(current_df['CD8 T cells']) < 0:
            cd8_low = cd8_low.append(row)
        else:
            cd8_high = cd8_high.append(row)
    

cd8_high['Status'] = df.apply(lambda row: bool(row.Status), axis=1)
cd8_low['Status'] = df.apply(lambda row: bool(row.Status), axis=1)
    
display(cd8_high)
cd8_low
```

### Survival Analysis Plotting

```python
time, survival_prob = kaplan_meier_estimator(df["Status"], df["Survival_in_days"])

plt.step(time, survival_prob, where="post")
plt.ylabel("Survival probablity")
plt.xlabel("Time")
```

```python
time_cd8_high, survival_prob_cd8_high = kaplan_meier_estimator(cd8_high["Status"], cd8_high["Survival_in_days"])
plt.step(time_cd8_high, survival_prob_cd8_high, where="post", label= 'Immune Active', color = 'maroon')

time_cd8_low, survival_prob_cd8_low = kaplan_meier_estimator(cd8_low["Status"], cd8_low["Survival_in_days"])
plt.step(time_cd8_low, survival_prob_cd8_low, where="post", label= 'Other', color = 'blue')

plt.legend(loc="upper right")
plt.ylabel("Survival probablity")
plt.xlabel("Time")
plt.savefig('TCGA_km.png')
```

```python
#logrank_test
results=logrank_test(time_cd8_high,time_cd8_low,event_observed_A=survival_prob_cd8_high, event_observed_B=survival_prob_cd8_low)
results.print_summary()
```
